```python
import math
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


plt.style.use('seaborn-v0_8')

output_dir = f"output/python/"

```

# Algorithm


```python



def hire_assistant(n , x ) :


    c = 0 # number of hires
    most_skilled_assistant = -1

    for i in range(1 , n + 1) :
        current_skill_assistant = random.randint(0 , int(x) )
        if current_skill_assistant > most_skilled_assistant :
            most_skilled_assistant = current_skill_assistant
            c += 1

    return c
```

# x = log2(n)

```python
avg_metrics = np.array([])
std_metrics = np.array([])

for n in range(1 , 1001) :
    clist = np.array([])


    for _ in range(10) :
        
        # n = 1000
        x = math.log(n,2.0) # x = log2(n)


        clist = np.append(clist , hire_assistant(n ,x))
        
    avg_metrics = np.append(avg_metrics , np.average(clist))
    std_metrics = np.append(std_metrics , np.std(clist))
        





        
```

```python

plt.plot( np.arange(1000) , std_metrics  , color = "blue" , label = "Std")
plt.title("Standard Deviation")
plt.savefig(f"{output_dir}/std_log2n.png")





plt.show()
```

```python
plt.plot( np.arange(1000) , avg_metrics , color = "red" , label = "Average")
plt.title("Average")
plt.savefig(f"{output_dir}/mean_log2n.png")

plt.show()
```

# x = n

```python
avg_metrics = np.array([])
std_metrics = np.array([])

for n in range(1 , 1001) :

    clist = np.array([])
    for _ in range(10) :
    
        x = n # x = n


        clist = np.append(clist , hire_assistant(n ,x))
        
    avg_metrics = np.append(avg_metrics , np.average(clist))
    std_metrics = np.append(std_metrics , np.std(clist))
```

```python

plt.plot( np.arange(1000) , std_metrics  , color = "blue" , label = "Std")
plt.title("Standard Deviation")
plt.savefig(f"{output_dir}/std_n.png")



plt.show()
```

```python
plt.plot( np.arange(1000) , avg_metrics , color = "red" , label = "Average")


plt.title("Average")
plt.savefig(f"{output_dir}/mean_n.png")

plt.show()
```

# x = n^3

```python
avg_metrics = np.array([])
std_metrics = np.array([])

for n in range(1 , 1001) :
    clist = np.array([])

    for _ in range(10) :
    
        x = n**3 # x = n^3


        clist = np.append(clist , hire_assistant(n ,x))
        
    avg_metrics = np.append(avg_metrics , np.average(clist))
    std_metrics = np.append(std_metrics , np.std(clist))
```

```python

plt.plot( np.arange(1000) , std_metrics  , color = "blue" , label = "Std")

plt.title("Standard Deviation")
plt.savefig(f"{output_dir}/std_n3.png")


plt.show()
```

```python
plt.plot( np.arange(1000) , avg_metrics , color = "red" , label = "Average")

plt.title("Average")
plt.savefig(f"{output_dir}/mean_n3.png")


plt.show()
```
